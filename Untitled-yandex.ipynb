Иниуиализация SS

```python
SS = StandardScaler()
```

```python
base_transformed = pd.DataFrame(SS.fit_transform(base), index=base.index, columns=base.columns)
```

```python
train_transformed = pd.DataFrame(SS.transform(train), index=train.index, columns=train.columns)
validation_transformed = pd.DataFrame(SS.transform(validation), index=validation.index, columns=validation.columns)
```

Масштабирование признаков, затем проверка корректности масштабирования.

```python
print(base_transformed.head())
print(train_transformed.head())
print(validation_transformed.head())
```

Определим размерность данных dimensions из датафрейма base. Индекс indexl2 создает индекс faiss при использовании l2 метрики.

```python
dimensions=base_transformed.shape[1]
indexl2=faiss.IndexFlatL2(dimensions)
```

Добавление данных в индекс

```python
indexl2.add(np.ascontiguousarray(base_transformed.values[:30000, :]).astype('float32'))
indexl2.ntotal
```

Создание словаря индексов, в котором сохраняется исходный индекс в датафрейме

```python
base_index = {k: v for k, v in enumerate(base_transformed.index.to_list())}
```

```python
D, I = indexl2.search(np.ascontiguousarray(train_transformed.values).astype('float32'), k)

acc = 0
for target, el in zip(train_target.values.tolist(), I.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 {100 * acc / len(I)}")
```

Метрика accuracy@5 3.004, так как с таким количеством и качеством данных не получилось использовать кластеры . Использование тренировочных данных из масштабированного датафрейма train_transformed. 

```python
D, I = indexl2.search(np.ascontiguousarray(validation_transformed.values).astype('float32'), k)

acc = 0
for target, el in zip(train_target.values.tolist(), I.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 на validation {100 * acc / len(I)}")
```
